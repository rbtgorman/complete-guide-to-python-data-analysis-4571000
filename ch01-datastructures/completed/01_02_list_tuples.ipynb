```python
squares = [1, 4, 9, 16, 25, 36, 49]
```

The convention to select a slice in Python is the same as for loops: the first index (0) is included, the last is not.

```python
squares[0:2]
```

```python
squares[:4]
```

```python
squares[3:]
```

```python
squares[:]
```

```python
squares[0:7:2]
```

```python
squares[-3:-1]
```

```python
# reverse the list!
squares[::-1]
```

```python
squares[2:4] = ['four', 'nine']
```

```python
squares
```

```python
del squares[4:6]
```

```python
squares
```
