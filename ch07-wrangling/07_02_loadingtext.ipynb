A simple pandas call does a good job with this.

```python
df = pd.read_csv('Planets.csv')
df
```

However we see that some columns have become strings (Python objects) instead of numbers, probably because of the commas; and that the "first visited" date is again a string rather than a date.

```python
df.info()
```

Luckily pandas has options for that. We specify the thousands separator, and tell pandas that the `FirstVisited` column holds dates.

```python
df = pd.read_csv('Planets.csv', thousands=',', parse_dates=['FirstVisited'])
df.info()
```

For consistency, we can require that the Diameter and MeanTemperature fields should be floats, and that `Rings` and `MagneticField` should be categories, since they have only two possible values.

```python
df = pd.read_csv('Planets.csv', thousands=',', parse_dates=['FirstVisited'],
                 dtype={'Diameter': np.float64, 'MeanTemperature': np.float64,
                        'Rings': 'category', 'MagneticField': 'category'})
df.info()
```

This is now a very clean, accurately encoded table!

Note that for this file, we would have had more trouble if we had loaded the table using the pyarrow backend, since in that case `read_csv` does not support `thousands` or `parse_dates`. We can however convert the DataFrame to pyarrow types after loading.

```python
df.convert_dtypes(dtype_backend='pyarrow').info()
```

pandas `read_csv` has many other useful options, such as skipping rows at the top and bottom, reading a fixed number of rows, selecting columns; there are options about parsing dates and floating numbers; we may provide a custom Python function to parse a specific column; and a lot more.

These options are mirrored in the writing function `to_csv`. We can specify the separator:

```python
df.to_csv('foo.csv', sep='\t')
print(open('foo.csv').read())
```

We can instruct pandas to write out null values instead of skipping them:

```python
df.to_csv('foo.csv', sep='\t', na_rep='NA')
print(open('foo.csv').read())
```

We can omit the index:

```python
df.to_csv('foo.csv', sep='\t', na_rep='NA', index=False)
print(open('foo.csv').read())
```

Or omit the header:

```python
df.to_csv('foo.csv', sep='\t', na_rep='NA', header=False, index=False)
print(open('foo.csv').read())
```

We can change the date format:

```python
df.to_csv('foo.csv', sep='\t', na_rep='NA', index=False, header=False, date_format='%d/%m/%Y')
print(open('foo.csv').read())
```
